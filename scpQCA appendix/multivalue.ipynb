```python
from itertools import product
import pandas as pd
import scpQCA
```

#### 3-value

```python
res = list(product([0,1,2],[0,1,2],[0,1,2],[0,1,2],[0,1,2],[0]))
data=pd.DataFrame(res)
data.columns=['A','B','C','D','E','OUTCOME']
data
```

```python
# A0*B0+B1*C1+C2*D2+D0*E0=OUTCOME
for i in range(len(data)):
    if (data.iloc[i]['A']==0 and data.iloc[i]['B']==0) or \
        (data.iloc[i]['B']==1 and data.iloc[i]['C']==1) or \
            (data.iloc[i]['C']==2 and data.iloc[i]['D']==2) or \
                (data.iloc[i]['D']==0 and data.iloc[i]['E']==0) or \
                    (data.iloc[i]['E']==1 and data.iloc[i]['A']==1):
        data.iloc[i]['OUTCOME']=1
data
```

```python
data.sum(axis=0)
```

```python
temp=data.sample(n=100,replace=True,axis=0)
temp
```

```python
temp.sum(axis=0)
```

```python
data=pd.read_csv(r'D:/code/scpQCA/paper code/multivalue/multivalue3.csv')
data
```

```python
feature_list=['A','B','C','D','E','OUTCOME']
test=scpQCA.scpQCA(data[feature_list],decision_name='OUTCOME',caseid='')
test.search_necessity(decision_label=1,feature_list=feature_list,consistency_threshold=0.9)
rules=test.candidate_rules(decision_label=1,feature_list=feature_list,consistency=0.8,cutoff=8,rule_length=2)
test.scp_truth_table(rules=rules,feature_list=feature_list,decision_label=1)
config,issue=test.greedy(rules=rules,decision_label=1,unique_cover=5)
test.cov_n_con(decision_label=1, configuration=config, issue_sets=issue)
print(config)
```
